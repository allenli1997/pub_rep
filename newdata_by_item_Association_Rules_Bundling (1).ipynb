```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SEP order sku detail.csv")
df
```

```python
df2 = df[['goods_ename','order_sn','domain']]
df2
```

```python
df3 = df2.groupby(['order_sn'])['goods_ename'].transform(lambda x : '^'.join(x))
```

```python
from pandas.core.groupby.ops import DataFrame
df3 = DataFrame(df3)
```

```python
df3 = df3.drop_duplicates()
```

```python
df3
```

```python
data = list(df3["goods_ename"].apply(lambda x:x.split("^") ))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df4 = pd.DataFrame(a_data,columns=a.columns_)
df4
```

```python
df5 = apriori(df4, min_support = 0.003, use_colnames = True)
df5
```

```python
df_ar = association_rules(df5, metric = "confidence", min_threshold = 0.01)
df_output = df_ar.sort_values(by='support', ascending=False)
df_output
```
